```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im=cv2.imread("parrot.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(im,cmap='grey')
plt.show()
```

```python
plt.hist(im.ravel(),256,range=[0,256])
plt.show()
```

```python
im_eq=cv2.equalizeHist(im)
```

```python
plt.hist(im_eq.ravel(),256,range=[0,256])
plt.show()
```

```python
img=cv2.imread("parrot.jpg",cv2.IMREAD_COLOR)
```

```python
img_hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
img_hsv[:,:,2]=cv2.equalizeHist(img_hsv[:,:,2])
```

```python
img_eq=cv2.cvtColor(img_hsv,cv2.COLOR_HSV2BGR)
```

```python
plt.hist(img_eq.ravel(),256,range=[0,256])
plt.show()
```

```python
plt.imshow(img_eq)
plt.title("equlized image")
plt.show()
```

```python
plt.imshow(img)
plt.title("original image")
plt.show()
```
